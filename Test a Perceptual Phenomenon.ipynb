(2) What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

Null Hypothesis-> mean time taken to name Congruent words is equal to mean time taken to name Incongruent words.  
H(0): μ(c)=μ(I)                           
Alternative Hypothesis-> mean time taken to name Congruent words is not equal to mean time taken to name Incongruent words.  
H(a): μ(c)!=μ(I)                                            
where:                                      
H(0)-> Null Hypothesis                                                        
H(a)-> Alternative Hypothesis                                                          
μ(c)-> Congruent population mean                                                     
μ(I)-> Incongruent population mean                                             
Statistical Test->  t-test=>we are conducting a t-test because here the sample size is small,the experiment is performed by same person therefore we analyse by doing a paired t-test.    
Justification-> I am taking a 2-tailed hypothesis where Null hypothesis is μ(c)=μ(I) and Alternative Hypothesis is              μ(c)!=μ(I),where μ(c) is mean time taken to name Congruent words and μ(I) is mean time taken to name Incongruent words.
t-test is used as if the p-value is less than the alpha(α),we can reject the null otherwise we have to accept the null.I will be assuming 5% error rate that makes α=0.05




(3) Report some descriptive statistics regarding this dataset. Include at least one measure of central tendency and at least one measure of variability. The name of the data file is 'stroopdata.csv'.

```python
# Perform the analysis here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df=pd.read_csv('stroopdata.csv')# reading the dataset
df.head()

mean_congruent=df.Congruent.mean() #calculate mean of Congurent
mean_incongruent=df.Incongruent.mean() #calculate mean of Incongurent

median_congruent=df.Congruent.median() #calculate median of Congurent
median_incongruent=df.Incongruent.median() #calculate median of Incongurent

sd_congruent=np.std(df['Congruent'],ddof=1) #calculate standard deviation of Congruent
sd_incongruent=np.std(df['Incongruent'],ddof=1) #calculate standard deviation of Incongruent

#printing the calculations done above
print('Mean of Congruent is =',mean_congruent)
print('Mean of Incongruent is =',mean_incongruent)
print('Median of Congruent is =',median_congruent)
print('Median of Incongruent is =',median_incongruent)
print('Standard Deviation of Congruent is =',sd_congruent)
print('Standard Deviation of Incongruent is =',sd_incongruent)

```

(5) Now, perform the statistical test and report the results. What is the confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

```python
# Perform the statistical test here
from scipy import stats
stats.ttest_rel(df['Congruent'],df['Incongruent'])
```

By the Hypothesis fact if pvalue <=  α then we reject the null. 
In this case we concluded that 4.1030005857111781e-08 <= 0.05 that is pvalue <= α therefore we reject the null and hence we can say that μ(c)!=μ(I) which means mean time taken to name Congruent words is not equal to mean time taken to name Incongruent words.Therefore we accept the Alternative Hypothesis.
